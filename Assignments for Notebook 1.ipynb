Solution:
##### First, we'll automatically generate some values: 

````
import random
#Generate 20 random numbers between 10 and 100
randomlist_x = random.sample(range(1, 100), 20)
print(randomlist_x)
````

These will be our "x" values.
Now it's your turn to generate another list of 20 random numbers between 20 and 100 to be our "y" values and name it *randomlist_y*:

````
#Generate 20 random numbers between 1 and 100
randomlist_y = random.sample(range(1, 100), 20)
print(randomlist_y)
````

Let's use the **sort()** function on our list of x values to sort the values in ascending order. This is a built- in  Python function. If you don't enter any additional parameters, it sorts the elements in the list in ascending order. This function changes our list directly, so there's no need to define a new sorted list. Add to the code to sort the values in the list you created. Also, use what you learned on notebook 1 to visualize the new version of your lists, to make sure the values are sorted correctly.

````
randomlist_x.sort()
randomlist_y.sort()

print(randomlist_x)
print(randomlist_y)
````

Let's import **numpy** and **matplotlib.pyplot** modules for our plot, and let's define our first list of numbers as our "x" values. Add your own code to set the second list as the "y" values:


````
import matplotlib.pyplot as plt
import numpy as np

x = randomlist_x
y = randomlist_y
````

Let's make a scatter plot with these values. Use the *scatter()* function of matplotlib. 
You can add a title to your graph witht he *title()* function, and label the axes with the xlabel() and ylabel() functions respectively.

````
plt.scatter(x,y, marker='o')
plt.title("Scatter Plot")
plt.xlabel('x values')
plt.ylabel("y values")
````

Now draw a trendline over the previous graph by defining a new variable as the 1-degree polynomial fit of your distribution. On numpy this function is called "polyfit", and the parameters should be x, y, and the degree of the polinomial you want to fit your distribution to.

In order to make a more undestandable graph, it's better to add "legends", to label everything properly. Add an additional "label" parameter to the scatter() function, and then use the  legend() function to show these legends on the graph.

````
plt.scatter(x, y, label="scatter")

z = np.polyfit(x, y, 1) # z gives the slope and intercept as it's output on a list
p = np.poly1d(z) #p will give the "y" values correponding to the x values.
plt.plot(x,p(x),"r-", label="trendline")

plt.title("Scatter Plot and trendline")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.show()
````

If we want to show the function equation, we can use: 


```
print("The curve fit for this function is y = {0} x + {1}.".format(z[0], z[1]))
```
"z" being the fit of the 1D polynomial. z[0] the slope of the adjusted function and z[1] the intercept.

For more information, please visit the [documentation](https://matplotlib.org/contents.html).


````
print("The curve fit for this function is y = {0:1.4E} x + {1:1.4E}.".format(z[0], z[1]))
````

Now let's draw some error bars. Use the *plt.errorbar()* function. You will need to add another variable: the y-error, which will be our linear regression minus the given "y" values. The parameters for your plt.errorbar() function will be these three variables.

````
plt.scatter(x, y, label="scatter")
plt.xlabel("x values")
plt.ylabel("y values")
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r-", label="trendline")
yerr=p(x)-y
plt.errorbar(x, p(x), yerr, fmt='.k',label="error")
plt.legend()
plt.show() 
print("The curve fit for this function is y = {0:1.4E} x + {1:1.4E}.".format(z[0], z[1]))
````
